```python
class Digit:
    
    def __init__(self, matrix, value):
        self.set_zipped_matrix(matrix)
        self.set_right_vector()
        self.set_left_vector()
        self.set_up_vector()
        self.set_down_vector()
        self.set_value(int(value))
    
    def __str__(self):
        return f"Water from up : {self.get_up_vector()} \nWater from left : {self.get_left_vector()} \nWater from down : {self.get_down_vector()}\nWater from right : {self.get_right_vector()} \nValue : {self.get_value()}\n\n"
    
    def set_zipped_matrix(self,matrix):
        self.zipped_matrix = self.zip_digit(matrix)
        
    def get_zipped_matrix(self):
        return self.zipped_matrix   
    
    def set_right_vector(self):
        self.right_vector = self.water_right(self.get_zipped_matrix())
    
    def get_right_vector(self):
        return self.right_vector
    
    def set_left_vector(self):
        self.left_vector = self.water_left(self.get_zipped_matrix())
    
    def get_left_vector(self):
        return self.left_vector
    
    def set_up_vector(self):
        self.up_vector = self.water_up(self.get_zipped_matrix())
    
    def get_up_vector(self):
        return self.up_vector
    
    def set_down_vector(self):
        self.down_vector = self.water_down(self.get_zipped_matrix())
    
    def get_down_vector(self):
        return self.down_vector
    
    def set_value(self, value):
        self.value = value
    
    def get_value(self):
        return self.value
        
    def zip_digit(self, matrix):
        old = matrix
        while np.all(matrix[0] == 0):       
            matrix = matrix[1:]
        while np.all(matrix[len(matrix)-1] == 0):       
            matrix = matrix[:len(matrix)-1]
        matrix = matrix.T
        while np.all(matrix[0] == 0):       
            matrix = matrix[1:]
        while np.all(matrix[len(matrix)-1] == 0):       
            matrix = matrix[:len(matrix)-1]
        matrix = matrix.T
        return matrix
        
    def water_left(self, matrix): #left calculation
        accumulated_left = 0
        for row in matrix:
            for col in row:
                if col == 0:
                    accumulated_left +=1
                else:
                    break
        return accumulated_left/matrix.size

    def water_right(self, matrix): #right calculation
        accumulated_right = 0
        for row in matrix:
            for col_num in range(len(row)-1,-1,-1):
                if row[col_num] == 0:
                    accumulated_right +=1
                else:
                    break
        return accumulated_right/matrix.size

    def water_up(self, matrix): #up calculation
        matrix = matrix.T
        accumulated_up = 0
        for row in matrix:
            for col in row:
                if col == 0:
                    accumulated_up +=1
                else:
                    break
        return accumulated_up/matrix.size

    def water_down(self, matrix):#down calculation
        matrix = matrix.T
        accumulated_down = 0
        for row in matrix:
            for col_num in range(len(row)-1,-1,-1):
                if row[col_num] == 0:
                    accumulated_down +=1
                else:
                    break
        return accumulated_down/matrix.size

    def save_to_csv(self, name, ):
        df = pd.DataFrame(self.zipped_matrix)
        df.to_csv(r"C:\Users\USER\Desktop\Durukan\Python\Digit-water\digits_excell\ " + name +r".csv")
```

```python
digit_data = data_frame
plt.rcParams.update({   "text.color": "white",
                        "axes.labelcolor": "white",
                        "axes.edgecolor": "white",
                        "xtick.color": "white",
                        "ytick.color": "white",
                        "grid.color": "white",
                        "figure.facecolor": "black",
                        "axes.facecolor": "black",
                        "legend.facecolor": "black",
                        "legend.edgecolor": "white",}
                     )


digit = np.zeros((500,650),dtype=int)


# Function to update the plot based on selected parameters
def update_plot(digit_estimated=None):
    global colorbar
    # Clear the previous plot
    ax.clear()
    # Clear color bar
    try:
        if colorbar:
            colorbar.remove()
            colorbar = None
    except:
        pass 
    
    # Get selected columns for x and y axes
    x_col = x_axis_var.get()
    y_col = y_axis_var.get()
    

    
    # Check if the selected columns are numeric
    if pd.api.types.is_numeric_dtype(digit_data[x_col]) and pd.api.types.is_numeric_dtype(digit_data[y_col]):
        scatter = ax.scatter(digit_data[x_col], digit_data[y_col], c=data_frame["value"], cmap = "tab10" , alpha=0.7)
        ax.set_title(f"Scatter Plot of {x_col} vs {y_col}")
        ax.set_xlabel(x_col)
        ax.set_ylabel(y_col)
        ax.set_facecolor("black")

        
    else:
        ax.text(0.5, 0.5, "Please select numeric columns", 
                ha='center', va='center', transform=ax.transAxes, color='red')
    if digit_estimated:
        match x_col:
            case "right_vector":
                highlighted_x = digit_estimated.get_right_vector()
            case "down_vector":
                highlighted_x = digit_estimated.get_down_vector()
            case "left_vector":
                highlighted_x = digit_estimated.get_left_vector()
            case "up_vector":
                highlighted_x = digit_estimated.get_up_vector()
            case "value":
                highlighted_x = digit_estimated.get_value()
        match y_col:
            case "right_vector":
                highlighted_y = digit_estimated.get_right_vector()
            case "down_vector":
                highlighted_y = digit_estimated.get_down_vector()
            case "left_vector":
                highlighted_y = digit_estimated.get_left_vector()
            case "up_vector":
                highlighted_y = digit_estimated.get_up_vector()
            case "value":
                highlighted_y = digit_estimated.get_value()
        ax.scatter(highlighted_x, highlighted_y, color="red", edgecolor="yellow", s=100, label="Highlighted Point", marker="^")  # Adjust size, color, and edge color as needed
    
    inset_ax = ax.inset_axes([0.9, 0.34, 0.015, 0.6])
    colorbar = fig.colorbar(scatter, cax=inset_ax,)
    colorbar.set_label("Value")  # Set the label for the color bar
    
    
    # Redraw the canvas
    canvas.draw()

def erase():
    global digit
    digit = np.zeros((500,650),dtype=int)
    C1.delete("all")
    print("canvas is erased!")

def paint(event):
    if -1<event.x<501 and -1<event.y<651:
        x,y=event.x,event.y
        for i in range(event.x - 5, event.x + 6):
            for k in range(event.y - 5, event.y + 6):
                if (499>i>0) and (649>k>0) :
                    digit[(i,k)] = 1
                    C1.create_line(i,k,i+1,k,fill="white")

def minkowski_distance(dataframe, digit,n):
    estimated_digit_right=digit.get_right_vector()
    estimated_digit_left=digit.get_left_vector()
    estimated_digit_down=digit.get_down_vector()
    estimated_digit_up=digit.get_up_vector()
    distances = []
    for index, row in dataframe.iterrows():
        rdistance = (abs(estimated_digit_right-row["right_vector"]))**n
        ldistance = (abs(estimated_digit_left-row["left_vector"]))**n
        ddistance = (abs(estimated_digit_down-row["down_vector"]))**n
        udistance = (abs(estimated_digit_up-row["up_vector"]))**n
        distance = (rdistance + ldistance + ddistance + udistance)**(1/n)
        distances.append(distance)
    return distances
    
def guess(data_frame,digit,k_var=20, n_var=3):
    k = k_var.get()
    n = n_var.get()
    #digit shown in the table
    digit_estimated = Digit(digit.T,-1)
    update_plot(digit_estimated)
    
    distances = minkowski_distance(data_frame, digit_estimated, n)
    neighbours = sorted(enumerate(distances),key=lambda x: x[1])[0:k]
    print("neigbours:", neighbours)
    neighbour_values = []
    for el in neighbours:
        neighbour_values.append(data_frame.iloc[int(el[0])]["value"])
    estimated_value = mode(neighbour_values) # get the most frequent neighbour value
    print("neighbours: ", neighbour_values)
    print("guess: ", estimated_value)
    C1.create_text(240,600,text=f"Result: {estimated_value}",font=("Helvetica Neue",20,"bold"),fill = "white", width=200)

def quit_application(event=None):
    root.quit()  # Quit the Tkinter event loop
    plt.close('all')
    root.destroy()

root = Tk()
root.title("Digit Recognizer")
root.geometry("1080x720")
root.configure(background="black")
C1 = Canvas(root, width=500, height=650, bg="black")
C2 = Canvas(root, width=500, height=650, bg="black")

B1 = Button(root, text="Erase",fg ="white", bg="black", font=("Arial",20), width=8, height=1, justify="center", command = erase)
B2 = Button(root, text="Guess",fg ="white", bg="black", font=("Arial",20), width=8, height=1, justify="center", command = lambda: guess(digit_data, digit, k_var, n_var)) # default k is 20

B1.place(x=80, y=660)
B2.place(x=280, y=660)

C1.place(x=0, y=0)
C2.place(x=580, y=0)

    


# Create the plot section 
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_facecolor("black")
fig.patch.set_facecolor('black')
fig.tight_layout()

# Integrate Matplotlib plot into Tkinter (on C2 canvas)
canvas = FigureCanvasTkAgg(fig, master=root)
canvas_widget = canvas.get_tk_widget()
canvas_widget.place(x=580, y=0, width=500, height=650)

# Create dropdowns for selecting x and y columns for plotting
x_axis_var = StringVar(value="right_vector")
y_axis_var = StringVar(value="left_vector")

x_axis_menu = OptionMenu(root, x_axis_var, *digit_data.columns)
x_axis_menu.config(bg="black", fg="white", font=("Arial",10), width=14, height=1, justify="center")
x_axis_menu.place(x=550, y=650)

y_axis_menu = OptionMenu(root, y_axis_var, *digit_data.columns)
y_axis_menu.config(bg="black", fg="white", font=("Arial",10), width=14, height=1, justify="center")
y_axis_menu.place(x=700, y=650)
print(digit_data.columns)
# KNN k=?
k_var = IntVar(value = 20)
k_menu = OptionMenu(root, k_var, *list(range(10,1005,5)))
k_menu.config(bg="black", fg="white", font=("Arial",10), width=14, height=1, justify="center")
k_menu.place(x=550, y=690)

# Minkowski distance n=?

n_var = IntVar(value=2)
n_menu = OptionMenu(root, n_var, *list(range(1,21)))
n_menu.config(bg="black", fg="white", font=("Arial",10), width=14, height=1, justify="center")
n_menu.place(x=700, y=690)

# Create an "Update Plot" button
update_button = Button(root, text="Update Plot", fg="white", bg="black", font=("Arial", 20), width=10, height=1, justify="center", command=update_plot)
update_button.place(x=870, y=660)


C1.bind("<B1-Motion>", paint)
root.bind("<q>", quit_application)
root.protocol("WM_DELETE_WINDOW", quit_application)
plt.close('all')


root.mainloop()
```
